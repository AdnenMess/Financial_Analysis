````python
import matplotlib.pyplot as plt
````

````python
import numpy as np

x = np.linspace(0,5,11)
y = x ** 2
````

````python
plt.plot(x, y, 'r')
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.show()
````

````python
plt.subplot(1, 2, 1)
plt.plot(x,y,'r')

plt.subplot(1, 2, 2)
plt.plot(y,x,'b')
plt.show()
````

````python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set title')
plt.show()
````

````python
import matplotlib.pyplot as plt

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Big Graph')
axes2.set_title('Small Graph')
plt.show()
````

**Matplotlib object-oriented method**

````python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First graph')

axes[1].plot(x, -y)
axes[1].set_title('Second Graph')
plt.tight_layout()
plt.show()
````

**figsize**

````python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12, 3))
axes[0].plot(x, y)
axes[1].plot(x, -y)
plt.tight_layout()
plt.show()
````

**Legend**

````python
fig, axes = plt.subplots()
axes.plot(x,x**2, label='X to the power of 2')
axes.plot(x,x**3, label='X to the power of 3')
axes.legend(loc=9)
plt.show()
````

**Legend outside chart**

````python
fig, axes = plt.subplots()
axes.plot(x,x**2, label='X to the power of 2')
axes.plot(x,x**3, label='X to the power of 3')
axes.legend(loc=9)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
````

**Color setting**

````python
fig, axes = plt.subplots()
axes.plot(x, y, color='#ff8c00', lw=3, alpha=0.5, ls=':')
plt.show()
````

````python
fig, axes = plt.subplots()
axes.plot(x, y, 'purple', lw=2, marker='o',  markersize=10,
          markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
plt.show()
````

````python
fig, axes = plt.subplots()
axes.plot(x, y,'g-.')
````

````python
fig, axes = plt.subplots(figsize=(8, 4))
axes.plot(x, y, 'purple', lw=2, ls='-')
axes.set_xlim([0,1])
axes.set_ylim([0,2])
plt.show()
````

**Special Graph**

````python
xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
````
