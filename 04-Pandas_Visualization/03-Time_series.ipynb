## Visualization time series

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

directory = Path.cwd().parent / "Files"
mcdon = pd.read_csv(directory / "mcdonalds.csv", index_col='Date', parse_dates=True)
```

```python
mcdon.head()
```

```python
mcdon['Adj. Close'].plot()
```

```python
mcdon['Adj. Volume'].plot(figsize=(12,4))
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-10', '2009-01-01'], ylim=(30,50))
```

```python
import matplotlib.dates as dates
```

```python
idx = mcdon.index
idx
```

```python
idx = mcdon.loc['2007-01-01':'2007-05-01'].index
idx
```

```python
stock = mcdon.loc['2007-01-01':'2007-05-01']['Adj. Close']
stock
```

```python
fig,ax = plt.subplots(figsize=(12,5))
ax.plot_date(idx, stock, '-')
ax.yaxis.grid(True)
ax.xaxis.grid(True)
fig.autofmt_xdate()
# Location
ax.xaxis.set_major_locator(dates.MonthLocator())
# Formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

plt.tight_layout()
plt.show()
```
