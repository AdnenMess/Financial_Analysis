## Integrated Data Visualization with Pandas

````python
import pandas as pd
import numpy as np
from pathlib import Path

directory = Path.cwd().parent / "Files"
df1 = pd.read_csv(directory / "df1", index_col=0)
df2 = pd.read_csv(directory / "df2")
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
````

````python
df1.head()
````

````python
df2.head()
````

**Styles**

   + ggplot
   + bmh
   + dark_background
   + fivethirtyeight

````python
df1['A'].hist(bins=30)
````

````python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
df1['A'].plot(kind='hist', bins=30)
````

````python
plt.style.use('bmh')
df1['A'].plot.hist(bins=30)
````

````python
plt.style.use('dark_background')
df1['A'].plot.hist(bins=30)
````

````python
plt.style.use('fivethirtyeight')
df1['A'].plot.hist(bins=30)
````

**Area**

````python
df2.plot.area(alpha=0.4)
````

**Bar**

````python
df2.plot.bar(stacked=True)
````

**Line**

````python
df1.plot.line(y='B', figsize=(12,4), lw=1)
````

**Scatter**

For more colors go to <https://matplotlib.org/2.0.2/users/colormaps.html>

````python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
````

````python
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
````

**Box**

````python
df2.plot.box()
````
