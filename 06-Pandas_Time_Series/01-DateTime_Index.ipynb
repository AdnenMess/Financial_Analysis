## Time Series with Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mtp
from datetime import datetime

my_year = 2022
my_month = 7
my_day = 30
my_hour = 13
my_minute = 30
my_second = 15
```

```python
my_date = datetime(my_year, my_month, my_day, my_hour, my_minute, my_second)
my_date
```

```python
my_date.month
```

```python
my_list = [datetime(2022, 1, 1), datetime(2022, 2, 1), datetime(2022, 3, 1), datetime(2022, 4, 1), datetime(2022, 5, 1), datetime(2022, 6, 1), datetime(2022, 7, 1), datetime(2022, 8, 1), datetime(2022, 9, 1), datetime(2022, 10, 1), datetime(2022, 11, 1), datetime(2022, 12, 1)]
```

```python
dt_ind = pd.DatetimeIndex(my_list)
dt_ind
```

```python
data = np.random.randn(12,2)
cols = ['a', 'b']
df = pd.DataFrame(data, dt_ind, cols)
df.head(3)
```

```python
df.index.max()
```

```python
df.index
```

```python
plt.style.use('seaborn-darkgrid')
fig,ax = plt.subplots(figsize=(12,5))
ax.plot_date(df.index, df['a'], '-', label='a')
ax.plot_date(df.index, df['b'], '-', label='b')
ax.yaxis.grid(True)
ax.xaxis.grid(True)
fig.autofmt_xdate()
# Location
ax.xaxis.set_major_locator(mtp.MonthLocator())
# Formatting
ax.xaxis.set_major_formatter(mtp.DateFormatter('\n%B-%Y'))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
plt.show()
```
